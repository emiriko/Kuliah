```python
import requests
import json
 
def fetch_wikidata(params):
    url = 'https://www.wikidata.org/w/api.php'
    try:
        return requests.get(url, params=params)
    except:
        return 'There was and error'
```

```python
query = 'Philadelphia'
 
# Which parameters to use
params = {
        'action': 'wbsearchentities',
        'format': 'json',
        'search': query,
        'language': 'en'
    }
 
# Fetch API
data = fetch_wikidata(params)
 
#show response as JSON
data = data.json()

dt = [data["search"][0]["label"], data["search"][0]["id"], data["search"][0]["description"]]
print(dt)
```

```python
df = pd.read_csv("Top_1000_Companies_Dataset.csv")
df.head()
```

```python
import csv

df['LeadInvestors'] = df['LeadInvestors'].str.split(', ')
df = df.explode('LeadInvestors')

df['keywords'] = df['keywords'].str.split(', ')
df = df.explode('keywords')

df.to_csv("new_df.csv", index=False, sep=',', quoting=csv.QUOTE_MINIMAL) 

df.head(15)
```
